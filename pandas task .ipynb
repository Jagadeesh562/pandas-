3. Add a column called in_stock which is True if quantity is greater than 0 and False if quantity equals 0.

```python
df['in_stocks'] = df.quantity.apply(lambda x : True if x>0 else False)
```

4. Create a column called  total_value that is equal to price multiplied by quantity.

```python
df['total_value']= df.apply(lambda row: row['quantity']*row['price'], axis= 1)
df.head()
```

5. create a new column called  full_description that has the complete description of each product. 
( product type and product description)

```python
df['full_description']= df.apply(lambda x: x['product_type']+' and '+x['product_description'], axis= 1)
df.head()
```
